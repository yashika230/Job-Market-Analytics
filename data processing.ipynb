```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Job Market Uncleaned.csv")
df
```

1. Basic Cleaning

```python
df.columns = df.columns.str.strip().str.lower()
df.drop_duplicates(inplace=True)
```

2. Handle Missing Values

```python
df.replace("NaN", np.nan, inplace=True)
```

```python
df.columns
```

```python
df.dropna(subset=["title", "company", "location"], inplace=True)
df.rename(columns={"title": "job_title"}, inplace=True)
df.columns
```

 3. Clean Job Title


```python
df["job_title_clean"] = df["job_title"].str.lower().str.strip()
```

```python
df["job_title_clean"] = df["job_title_clean"].replace({
    "sde": "software engineer",
    "developer": "software engineer"})
```

4. Clean Location


```python
df["location_clean"] = df["location"].str.strip().str.upper()
```

```python
df = df[df["location_clean"] != "UNITED STATES"]
df["location_clean"]
```

 5. Salary Cleaning


```python
df["sal_low"] = pd.to_numeric(df["sal_low"], errors="coerce")
df["sal_high"] = pd.to_numeric(df["sal_high"], errors="coerce")
```

```python
df["avg_salary"] = (df["sal_low"] + df["sal_high"]) / 2
```

6. Skills Cleaning 


```python
df.columns
```

```python
import pandas as pd
from collections import Counter

SKILLS_TO_EXTRACT = [
    'python', 'java', 'javascript', 'go', 'sql', 'aws', 'docker', 
    'kubernetes', 'react', 'machine learning', 'git', 'linux'
]

df = pd.read_csv('Job Market Uncleaned.csv')

def find_skills(description):
    if pd.isna(description):
        return []
    
    description_lower = str(description).lower()
    found_skills = []
    
    for skill in SKILLS_TO_EXTRACT:
        if skill in description_lower:
            found_skills.append(skill.title())
    
    return list(set(found_skills))

df['skills'] = df['description'].apply(find_skills)

df = df[df['skills'].apply(len) > 0]
```

```python
df["skills"] = df["skills"].str.replace(r"[\[\]']", "", regex=True)
```

```python
df.columns
```

7. Industry Cleaning


```python
df["Industries"] = df["Industries"].str.strip().str.lower()
```

```python
df["Industries"] = df["Industries"].replace({
    "Information technology services": "Information technology",
    "It services": "Information technology"})
```

8. Final Clean Dataset


```python
df.to_csv("Job Market cleaned.csv", index=False)
```
